```python
import pandas as pd
import glob
import numpy as np
from operator import truediv
```

```python
def get_X_Y(df):
    """
    Get big X and big Y
    """
    X = df.iloc[:,1:5].values
    Y = df.iloc[:,0].values
    
    return X,Y

def get_center(df,param):
    """
    Get the center of x or y
    This function serves as a sub-function for get_center_division()
    :param: Can be 'x' or 'y', specify which center to be calculated
    """
    if param == 'x':
        list_x_center = []
        for i in range(0, len(df)):
            x_center = (X_classification[i][2] - X_classification[i][0])/2 # (xmax - xmin)/2
            list_x_center.append(x_center)
        return list_x_center
    elif param == 'y':
        list_y_center = []
        for i in range(0, len(df)):
            y_center = (X_classification[i][3] - X_classification[i][1])/2 # (ymax - ymin)/2
            list_y_center.append(y_center)
        return list_y_center
    else:
        return None
    
def get_center_division(df,center):
    """
    Get the x divided by width or y divided by height
    :param: Can be 'x' or 'y', specify which center to be calculated
    """
    if center == 'x':
        the_center = get_center(df,center)
        width = df['width'].values
        xDVwidth = list(map(truediv, the_center, width)) 

        return xDVwidth
    elif center == 'y':
        the_center = get_center(df,center)
        height = df['height'].values
        yDVheight = list(map(truediv, the_center, height)) 

        return yDVheight
    else:
        return None

def get_maxmin_subtract(df,param):
    """
    Get the center of x or y
    This function serves as a sub-function for get_subtracted_division()
    :param: Can be 'x' or 'y', specify which center to be calculated
    """
    if param == 'x':
        list_x_subtract = []
        for i in range(0, len(df)):
            list_x_subtract.append(X_classification[i][2] - X_classification[i][0]) # xmax - xmin
        return list_x_subtract
    elif param == 'y':
        list_y_subtract = []
        for i in range(0, len(df)):
            list_y_subtract.append(X_classification[i][3] - X_classification[i][1]) # ymax-ymin
        return list_y_subtract
    else:
        return None
    
def get_subtracted_division(df,center):
    """
    Get the x divided by width or y divided by height
    :param: Can be 'x' or 'y', specify which center to be calculated
    """
    if center == 'x':
        the_center = get_maxmin_subtract(df,center)
        width = df['width'].values
        xDVwidth = list(map(truediv, the_center, width)) 

        return xDVwidth
    elif center == 'y':
        the_center = get_maxmin_subtract(df,center)
        height = df['height'].values
        yDVheight = list(map(truediv, the_center, height)) 

        return yDVheight
    else:
        return None

def get_detected_class(df1,df2):
    """
    Get the detected class from df2 with if equal coordinates
    :param df1: the classification dataframe
    :param df2: the detection dataframe
    :return: the list of featured_class taken from df2
    """
    result = []
    for i in range(0,len(df1)):
        if X_classification[i].all() == X_detection[i].all():
            result.append(df2.iloc[i][0])
    return result

            
path = 'csv_classification/' 
all_files = glob.glob(path + "/*.csv")

li = []
i = 0
for filename in all_files:
    i += 1
    if i < 10:
        a_df = pd.read_csv(filename)
        print(a_df)
        li.append(a_df)
    
df_classification = pd.concat(li, axis=0, ignore_index=True)

df_classification

# df_classification = pd.read_csv('csv_classification/1609644373_etH3C.jpg@product_block_13.jpg.csv')
# df_classification
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier =  RandomForestClassifier(n_estimators=10, max_depth=20, min_samples_split=8, max_features=1,
                                 class_weight='balanced', min_samples_leaf=1)
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
